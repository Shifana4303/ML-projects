```python
X = df.drop(['Name','Surname'], axis=1)
y = df['Result']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()

mm.fit(x_train)
x_train=mm.transform(x_train)
x_train
```

```python
x_test = mm.transform(x_test)
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_predict=knn.predict(x_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_predict)*100)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,y_predict))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
print(ConfusionMatrixDisplay.from_predictions(y_test,y_predict))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predict))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='entropy', random_state=42)
dt.fit(x_train, y_train)
```

```python
y_predict = dt.predict(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
```

```python
svm_clf = SVC(kernel='rbf', probability=True, random_state=42)
knn_clf = KNeighborsClassifier(n_neighbors=5)
dt_clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
```

```python
voting_clf = VotingClassifier(
    estimators=[('svm', svm_clf), ('knn', knn_clf), ('dt', dt_clf)],
    voting='soft'
)
```

```python
voting_clf.fit(x_train, y_train)
```

```python
y_pred_voting = voting_clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(accuracy_score(y_test, y_pred_voting))
```

```python
print(confusion_matrix(y_test, y_pred_voting))
```

```python
print(classification_report(y_test, y_pred_voting))
```

```python
voting_hard = VotingClassifier(
    estimators=[('svm', svm_clf), ('knn', knn_clf), ('dt', dt_clf)],
    voting='hard'
)
```

```python
voting_soft = VotingClassifier(
    estimators=[('svm', svm_clf), ('knn', knn_clf), ('dt', dt_clf)],
    voting='soft'
)
```

```python
voting_hard.fit(x_train, y_train)
voting_soft.fit(x_train, y_train)
```

```python
y_pred_hard = voting_hard.predict(x_test)
y_pred_soft = voting_soft.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

acc_hard = accuracy_score(y_test, y_pred_hard)
acc_soft = accuracy_score(y_test, y_pred_soft)

print(f"Hard Voting Accuracy: {acc_hard:.4f}")
print(f"Soft Voting Accuracy: {acc_soft:.4f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(accuracy_score(y_test, y_pred_voting))
```

```python
print(confusion_matrix(y_test, y_pred_voting))
```

```python
print(classification_report(y_test, y_pred_voting))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_voting))
```

```python
svm_clf.fit(x_train, y_train)
knn_clf.fit(x_train, y_train)
dt_clf.fit(x_train, y_train)

svm_acc = accuracy_score(y_test, svm_clf.predict(x_test)) * 100
knn_acc = accuracy_score(y_test, knn_clf.predict(x_test)) * 100
dt_acc  = accuracy_score(y_test, dt_clf.predict(x_test)) * 100
voting_acc = accuracy_score(y_test, y_pred_voting) * 100

import pandas as pd

results = pd.DataFrame({
    'Model': ['SVM', 'KNN', 'Decision Tree', 'Voting Classifier'],
    'Accuracy (%)': [svm_acc, knn_acc, dt_acc, voting_acc]
})

print(results.to_string(index=False))
```

# Random forest

```python
import pandas as pd

df=pd.read_csv("/content/diabetes.csv")
df.info()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python

X=df.iloc[:,0:8]
print(X)
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

mm=MinMaxScaler()
mm.fit(x_train)
x_train=mm.transform(x_train)
```

```python
x_train
```

```python
x_test=mm.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_predict=knn.predict(x_test)
```

```python
y_predict
```

```python
y_test.values
```

```python
new=knn.predict(mm.transform([[2,84,0,0,0,0.0,0.304,21]]))
```

```python
new
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_predict)*100)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_predict))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10)
rf.fit(x_train,y_train)
y_predict=rf.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print(accuracy_score(y_test,y_predict)*100)
print(confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
```
