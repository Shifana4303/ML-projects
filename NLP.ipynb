```python
import pandas as pd
df=pd.read_csv("/content/news-data.csv",nrows=2000)          #to reduce the no. of rows
df.head()
```

```python
news=df['headline_text']
```

```python
news=news.str.replace("[^a-z A-Z 0-9]+","",regex=True)
```

```python
news
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
wn=WordNetLemmatizer()
```

```python
news=news.apply(lambda x:" ".join([wn.lemmatize(i.lower())for i in tk.tokenize(x)]))
```

```python
news
```

```python
import nltk

from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words("english")
```

```python
news = news.apply(lambda text: " ".join([word for word in text.split() if word not in sw]))
```

```python
news
```

```python
news=news.str.replace("[^a-z A-Z 0-9]+","",regex=True)
```

```python
news
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x=tfidf.fit_transform(news)
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.show()
```

```python
km=KMeans(n_clusters=6,init="k-means++",random_state=42)
y_predict=km.fit_predict(x)
```

```python
y_predict
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x,y_predict)
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=42)
x_new,y_new=sm.fit_resample(x,y)
```

```python
from collections import Counter
print(Counter(y))
print(Counter(y_new))
```

```python
print(Counter(y))
print(Counter(y_ud))
```
